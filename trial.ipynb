```python
import os
import re
import pandas as pd
from PyPDF2 import PdfReader
from datetime import datetime

def parse_pdf(file_path):
    with open(file_path, 'rb') as file:
        pdf_reader = PdfReader(file)
        text = ""
        for page_num in range(len(pdf_reader.pages)):
            text += pdf_reader.pages[page_num].extract_text()
    return text

# Example usage to check the parse_pdf function
sample_pdf_path = r'E:\Projects\Victor-Files\annual_reports\Absa_group\2022-Absa-Group-limited-Environmental-Social-and-Governance-Data-sheet.pdf'  # Replace with the path to a sample PDF file

try:
    extracted_text = parse_pdf(sample_pdf_path)
    print("Text extracted from the PDF:")
    print(extracted_text)
except Exception as e:
    print(f"An error occurred: {e}")
```
